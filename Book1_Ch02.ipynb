```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
```

```python
loadfile = './data/sst_OISST.npz'
sst_dataset = np.load(loadfile)
```

```python
sst_dataset.files
```

```python
# sea surface temperature
sst = sst_dataset['sst']

#
lon2 = sst_dataset['lon2']

#
lat2 = sst_dataset['lat2']

# yaer
y = sst_dataset['y']

# month
m = sst_dataset['m']
```

### draw

```python
draw_y = 1997
draw_m = 12

vmin = -5
vmax = 35
vint = 5

cm = plt.get_cmap('seismic')

cs = plt.contourf(lon2, lat2, \
    np.squeeze(sst[:, :, (y==draw_y)*(m==draw_m)]), \
    cmap=cm, norm=Normalize(vmin=vmin, vmax=vmax), \
    levels=np.arange(vmin, vmax*vint, vint), extend='both')

plt.colorbar(cs)
plt.xlabel('Longitude')
plt.xlabel('Latitude')
plt.xlim(0, 360)
plt.ylim(-90, 90)

title = str(draw_y)
plt.title(title)

plt.show()
```
